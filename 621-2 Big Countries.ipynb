## Задача "Big Countries"

**Таблица:** `World`

| Column Name | Type    |
|-------------|---------|
| name        | varchar |
| continent   | varchar |
| area        | int     |
| population  | int     |
| gdp         | bigint  |

- **name** является первичным ключом (колонка с уникальными значениями) для этой таблицы.
- Каждая строка в таблице содержит информацию о названии страны, континенте, к которому она относится, площади, населении и значении ВВП.

### Условия:

Страна считается "большой", если:
- Её площадь составляет не менее трёх миллионов км² (т.е. `area >= 3000000`), или
- Её население составляет не менее двадцати пяти миллионов человек (т.е. `population >= 25000000`).

### Задача:

Напишите решение для вывода имени, населения и площади больших стран.

**Формат вывода должен соответствовать следующему примеру.**

### Пример:

**Входные данные:**
Таблица `World`:

| name        | continent | area    | population | gdp          |
|-------------|-----------|---------|------------|--------------|
| Afghanistan | Asia      | 652230  | 25500100   | 20343000000  |
| Albania     | Europe    | 28748   | 2831741    | 12960000000  |
| Algeria     | Africa    | 2381741 | 37100000   | 188681000000 |
| Andorra     | Europe    | 468     | 78115      | 3712000000   |
| Angola      | Africa    | 1246700 | 20609294   | 100990000000 |

**Ожидаемый результат:**

| name        | population | area    |
|-------------|------------|---------|
| Afghanistan | 25500100   | 652230  |
| Algeria     | 37100000   | 2381741 |


```python
import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    world = world.query('area >= 3000000 or population >= 25000000')[['name', 'population', 'area']]
    return world
```

```python
import pandas as pd

def big_countries(world: pd.DataFrame) -> pd.DataFrame:
    return world[
        (world['area'] >= 3000000) |
        (world['population'] >= 25000000)
    ][['name', 'population', 'area']]
```
