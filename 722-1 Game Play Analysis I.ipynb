```python
import pandas as pd

def game_analysis(activity: pd.DataFrame) -> pd.DataFrame:
    """
    Функция находит первую дату входа для каждого игрока.

    :param activity: DataFrame с колонками player_id, device_id, event_date, games_played
    :return: DataFrame с колонками player_id и first_login
    """
    # Группируем данные по player_id и находим минимальную дату в event_date
    first_login = activity.groupby('player_id', as_index=False)['event_date'].min()

    # Переименовываем колонку event_date в first_login
    first_login.rename(columns={'event_date': 'first_login'}, inplace=True)

    return first_login

# Пример входных данных
data = {
    'player_id': [1, 1, 2, 3, 3],
    'device_id': [2, 2, 3, 1, 4],
    'event_date': ['2016-03-01', '2016-05-02', '2017-06-25', '2016-03-02', '2018-07-03'],
    'games_played': [5, 6, 1, 0, 5]
}

# Создание DataFrame
activity = pd.DataFrame(data)

# Преобразование event_date в формат datetime для корректной работы
activity['event_date'] = pd.to_datetime(activity['event_date'])

# Вызов функции
result = game_analysis(activity)

# Печать результата
print(result)
```
