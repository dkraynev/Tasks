```python
import pandas as pd

def find_managers(employee: pd.DataFrame) -> pd.DataFrame:
    managers = employee.groupby('managerId').size().reset_index(name='count')
    managers = managers[managers['count'] >= 5]
    result = managers.merge(
        employee,
        left_on = 'managerId',
        right_on = 'id',
        how = 'inner'
    )[['name']]

    return result
```
