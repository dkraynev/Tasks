```python
import pandas as pd

def find_patients(patients: pd.DataFrame) -> pd.DataFrame:
    patients = patients[patients['conditions'].str.contains(r'(^|\s)DIAB1')]
    return patients
```

```python
import pandas as pd

def find_patients(patients: pd.DataFrame) -> pd.DataFrame:
    return patients[
        patients['conditions'].str.startswith('DIAB1') |
        patients['conditions'].str.contains(' DIAB1')
    ]
```
