```python
import pandas as pd
```

```python
def meltTable(report):
    report = pd.melt(report, id_vars = ['product'], var_name = 'quarter', value_name = 'sales')

    return report
```

```python
# Пример DataFrame report
data = {
    'product': ['Umbrella', 'SleepingBag'],
    'quarter_1': [417, 800],
    'quarter_2': [224, 936],
    'quarter_3': [379, 93],
    'quarter_4': [611, 875]
}

report = pd.DataFrame(data)
print(report)
print()

# Вызываем функцию для преобразования таблицы
result = meltTable(report)
print(result)
```
