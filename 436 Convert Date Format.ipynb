```python
import pandas as pd

# Пример данных
data = {
    'day': ['2022-04-12', '2021-08-09', '2020-06-26']
}
days = pd.DataFrame(data)

# Преобразование столбца в формат datetime, если это еще не сделано
days['day'] = pd.to_datetime(days['day'])

# Преобразование даты в требуемый формат
days['formatted_day'] = days['day'].dt.strftime('%A, %B %-d, %Y')

# Вывод результата
print(days[['formatted_day']])
```

```python
import pandas as pd

def convert_date_format(days: pd.DataFrame) -> pd.DataFrame:
    # Преобразование столбца day в формат datetime, если это ещё не сделано
    days['day'] = pd.to_datetime(days['day'])

    # Преобразование даты в требуемый формат
    days['day'] = days['day'].dt.strftime('%A, %B %-d, %Y')

    # Возвращаем только изменённый столбец day
    return days[['day']]

# Пример использования
data = {
    'day': ['2022-04-12', '2021-08-09', '2020-06-26']
}
days = pd.DataFrame(data)

# Вызов функции
formatted_days = convert_date_format(days)
print(formatted_days)
```
